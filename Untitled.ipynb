```python
#KNN ( K- NEAREST NEIGHBOURS)
#daha önce elde edilen noktalara bakarak yeni bir nokta geldiğinde onun hangi grupta olduğunu bulmak için kullanılır
#bir sınıflandırma modeli diyebiliriz
#en yakın K tane komşu nokta bulunur
#en çok hangi gruptan komşu nokta varsa yeni gelen noktanın o gruptan olduğu kabul edilir
#uzaklıkları ölçerken oklid mesafesi kullanır.
# K değeri yeni noktanın çevresindeki komşu noktalardan kaç tanesinin seçilip algoritmada yer alacağını verir.

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

#1= şeker hastasi
#0= sağlıklı
```

```python
data= pd.read_csv("diabetes.csv")
data
```

```python
# filtering işlemi ile şeker hastası ve sağlıklı olan hastaları ayrıştırıyoruz.
şeker_hastası= data[data.Outcome==1]
sağlıklı=data[data.Outcome==0]
sağlıklı
```

```python
#şimdilik glikoz değerlerine bakarak bir scatter plot çizdirelim.
plt.scatter(şeker_hastası.Age, şeker_hastası.Glucose, color= "red", label= "şeker hastası", alpha= 0.5)
plt.scatter(sağlıklı.Age, sağlıklı.Glucose, color= "green", label="sağlıklı", alpha= 0.5)
plt.xlabel("age")
plt.ylabel("glucose")
plt.legend()
plt.show()

#alpha değeri noktaların renk yoğunluğu ile ilgili, değiştirilebilir
```

```python
#x ve y'i belirleyelim
y= data.Outcome.values
x_ham = data.drop(["Outcome"], axis= 1)
print(x_ham)
# Outcome sütununu(dependent variable) çıkarıp sadece independent variables bırakıyoruz.Çüknü KNN algoritması x değerleri içerisinde gruplandırma yapacak
#axis değeri 0 ise yatay 1 ise dikey sıralanır yani o iken satırlar 1 iken sütunlar üzerinde işlem gerçekleştirilir
```

```python
#Normalizasyon, belirli bir veri kümesi için en basit yapıyı bulmakla ilgilidir. ve veri tutarsızlıklarını önlemek için normalleştirme yapılır.
#algoritmalar çalışma sistemlerine göre farklı normalizasyon tekniklerine ihtiyaç duyarlar
#KNN,SVM,K-Means gibi mesafeye dayalı(distance based) algoritmalar feature scalling yöntemine ihtiyaç duyarlar. Min-Max Scalling olarak da bilinir.
# min- max scalling için x_ham_veri içerisindeki değerleri sadece 0 ve 1 arasında olacak şekilde hepsini güncelliyoruz
# Eğer bu şekilde normalization yapmazsak yüksek rakamlar küçük rakamları ezer ve KNN algoritmasını yanıltabilir!
x = (x_ham - np.min(x_ham))/(np.max(x_ham)-np.min(x_ham))

print("Normalization öncesi ham veriler:\n")
print(x_ham.head())

print("Normalization sonrası kullanacağımız veriler:\n")
print(x.head())
```

```python
# train datamız ile test datamızı ayırıyoruz
# train datamızı modeli eğitmek için kullanacağız
# test datamızı ise modeli test etmek için kullanacağız

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state=1)
#test_size=0.2 verinin %20sinin test olarak ayrılacağını ve random_state=1 ise rastgele seçileceğini ifade eder.

# knn modelimizi oluşturuyoruz ve eğitiyoruz.
knn = KNeighborsClassifier(n_neighbors = 3) # n_neighbors = k
knn.fit(x_train,y_train)
#prediction ile sonuçların performansına bakalım
prediction = knn.predict(x_test)
print("K=3 için Test verilerimizin doğrulama testi sonucu ", knn.score(x_test, y_test))
```

```python
#%78 başarılı çalışıyor sonucuna varırız. Çok kötü olmamakla beraber çok iyi bir performans da diyemeyiz.
# k kaç olmalı ?
# en iyi k değerini belirleyelim..
sayac = 1
for k in range(1,11):
    knn_yeni = KNeighborsClassifier(n_neighbors = k)
    knn_yeni.fit(x_train,y_train)
    print(sayac, "  ", "Doğruluk oranı: %", knn_yeni.score(x_test,y_test)*100)
    sayac += 1
```

```python
knn = KNeighborsClassifier(n_neighbors = 7) # en iyi K değerini 7 olarak bulmuştuk
knn.fit(x_train,y_train)
#prediction ile sonuçların performansına bakalım
prediction = knn.predict(x_test)
print("K=7 için Test verilerimizin doğrulama testi sonucu ", knn.score(x_test, y_test))
```

```python
# model performansını artırmak için eğitim verisini büyütmek de bir seçenektir. Daha fazla veriyle beslenen model daha iyi eğitilir ve performansı artar.
# bunun için test size değerini düşürerek train size'ı artırabiliriz.
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.1,random_state=1)
# knn modelimizi oluşturuyoruz ve eğitiyoruz.
knn = KNeighborsClassifier(n_neighbors = 3) # n_neighbors = k
knn.fit(x_train,y_train)
#prediction ile sonuçların performansına bakalım
prediction = knn.predict(x_test)
print("K=3 için Test verilerimizin doğrulama testi sonucu ", knn.score(x_test, y_test))
```

```python
#görüldüğü üzere performans %83 çıktı. Bu performans çeşitli yöntemlerle daha da iyileştirilerek değer yükseltilebilir.
```
